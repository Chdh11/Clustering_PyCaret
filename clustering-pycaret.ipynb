```python
from pycaret.clustering import *
```

```python
#Clustering Techniques Used
clus_tech=['kmeans','hclust','dbscan']
#Pre-processing Techniques Used
pre_processing_tech=['none','norm','transform','pca','tn','tnpca']

def clustering(technique,prepro):
    #No Data Preprocessing
    if prepro=='none':          
        setup(data = df, verbose=False)
        print("No Data Preprocessing")
        x = create_model('kmeans')
        plot_model(x, plot = 'elbow')
        print("To different number of clusters:")
    #Normalization
    elif prepro=='norm':        
        setup(data = df, normalize = True, normalize_method = 'zscore', verbose=False)
        print("Using Normalization")
        x = create_model('kmeans')
        plot_model(x, plot = 'elbow')
        print("To different number of clusters:")
    #Transformation
    elif prepro=='transform':   
        setup(data = df, transformation = True, transformation_method = 'yeo-johnson', verbose=False)
        print("Using Transformation")
        x = create_model('kmeans')
        plot_model(x, plot = 'elbow')
        print("To different number of clusters:")
    #Pricipal Component Analysis
    elif prepro=='pca':         
        setup(data = df, pca = True, pca_method = 'linear', verbose=False)
        print("Using Pricipal Component Analysis")
        x = create_model('kmeans')
        plot_model(x, plot = 'elbow')
        print("To different number of clusters:")
    #Transformation + Normalization
    elif prepro=='tn':          
        setup(data = df, transformation = True, normalize = True,
              normalize_method = 'zscore', transformation_method = 'yeo-johnson', verbose=False)
        print("Using Transformation + Normalization")
        x = create_model('kmeans')
        plot_model(x, plot = 'elbow')
        print("To different number of clusters:")
    #Transformation + Normalization + PCA
    elif prepro=='tnpca':       
        setup(data = df, transformation = True, normalize = True, pca = True,
              normalize_method = 'zscore',
              transformation_method = 'yeo-johnson',
              pca_method = 'linear', verbose=False)
        print("Using Transformation + Normalization + PCA")
        x = create_model('kmeans')
        plot_model(x, plot = 'elbow')
        print("To different number of clusters:")
    else:
        print("Incorrect Pre-processing technique!!")
        
    print("For Cluster = 3")
    x1 = create_model(technique, num_clusters = 3)

    print("For Cluster = 4")
    x2 = create_model(technique, num_clusters = 4)

    print("For Cluster = 5")
    x3 = create_model(technique, num_clusters = 5)

    print("For Cluster = 6")
    x4 = create_model(technique, num_clusters = 6)
```

### K-means Clustering

```python
for i in pre_processing_tech:
    clustering('kmeans',i)
```

### Agglomerative Clustering (Heirarchical)

```python
for i in pre_processing_tech:
    clustering('hclust',i)
```

### Density-Based Spatial Clustering

```python
for i in pre_processing_tech:
    clustering('dbscan',i)
```
